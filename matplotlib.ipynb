```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
a = np.linspace(0,5,11)
```

```python
b = a ** 2
```

```python
plt.plot(a,b)
plt.xlabel('number')
plt.ylabel('squares')
plt.title('power')
```

```python
x,y = a,b
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(y,x)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,.6,.6])
axes = fig.add_axes([0.2,0.2,.4,.4])
axes = fig.add_axes([0.3,0.3,.4,.4])
axes.plot(x,y)
```

```python

plt.show()
```

```python
plt.figure()
plt.plot(x,y,label='x')
plt.plot(y,x,label='y')
plt.legend(loc='upper left')
```

```python
fig = plt.figure(figsize=(7,3))
axes = fig.add_axes([0.1,0.1,1,1])
axes.plot(x,y)
```
