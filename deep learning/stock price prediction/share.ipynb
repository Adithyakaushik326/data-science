```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```
df = pd.read_excel('drive/My Drive/colab-data/Train_dataset.xlsx')
```

**filling Null values**
group by industry and fill its respective mean


```
def computeVWAP(cols):
  industry = cols[0]
  vwap= cols[1]
  if pd.isnull(vwap):
    return df[df['Industry']==industry]['VWAP'].mean()
  else:
    return vwap
  
def computeGi(cols):
  industry = cols[0]
  GI= cols[1]
  if pd.isnull(GI):
    return df[df['Industry']==industry]['General Index'].mean()
  else:
    return GI
  
def computeNav(cols):
  industry = cols[0]
  nav= cols[1]
  if pd.isnull(nav):
    return df[df['Industry']==industry]['NAV'].mean()
  else:
    return nav

def computepe(cols):
  industry = cols[0]
  nav= cols[1]
  if pd.isnull(nav):
    return df[df['Industry']==industry]['P/E Ratio'].mean()
  else:
    return nav

def computevt(cols):
  industry = cols[0]
  nav= cols[1]
  if pd.isnull(nav):
    return df[df['Industry']==industry]['Volumes Traded'].mean()
  else:
    return nav

def computeit(cols):
  industry = cols[0]
  nav= cols[1]
  if pd.isnull(nav):
    return df[df['Industry']==industry]['Inventory Turnover'].mean()
  else:
    return nav

def computec(cols):
  industry = cols[0]
  nav= cols[1]
  if pd.isnull(nav):
    return df[df['Industry']==industry]['Covid Impact (Beta)'].mean()
  else:
    return nav

def computete(cols):
  industry = cols[0]
  nav= cols[1]
  if pd.isnull(nav):
    return df[df['Industry']==industry]['Tracking Error'].mean()
  else:
    return nav

def computed(cols):
  industry = cols[0]
  nav= cols[1]
  if pd.isnull(nav):
    return df[df['Industry']==industry]['Dollar Exchange Rate'].mean()
  else:
    return nav

def computepc(cols):
  industry = cols[0]
  nav= cols[1]
  if pd.isnull(nav):
    return df[df['Industry']==industry]['Put-Call Ratio'].mean()
  else:
    return nav

def computepb(cols):
  industry = cols[0]
  nav= cols[1]
  if pd.isnull(nav):
    return df[df['Industry']==industry]['P/B Ratio'].mean()
  else:
    return nav
    
df['P/B Ratio']=df[['Industry','P/B Ratio']].apply(computepb,axis=1)
df['Put-Call Ratio']=df[['Industry','Put-Call Ratio']].apply(computepc,axis=1)
df['Dollar Exchange Rate']=df[['Industry','Dollar Exchange Rate']].apply(computed,axis=1)
df['Tracking Error']=df[['Industry','Tracking Error']].apply(computete,axis=1)
df['Covid Impact (Beta)']=df[['Industry','Covid Impact (Beta)']].apply(computec,axis=1)
df['Inventory Turnover']=df[['Industry','Inventory Turnover']].apply(computeit,axis=1)
df['Volumes Traded']=df[['Industry','Volumes Traded']].apply(computevt,axis=1)
df['P/E Ratio']=df[['Industry','P/E Ratio']].apply(computepe,axis=1)
df['NAV']=df[['Industry','NAV']].apply(computeNav,axis=1)
df['General Index']=df[['Industry','General Index']].apply(computeGi,axis=1)
df['VWAP']=df[['Industry','VWAP']].apply(computeVWAP,axis=1)

```

```
df.isnull().sum()
```

```
x = df[['Index', 'Industry', 'VWAP', 'General Index', 'NAV',
       'P/E Ratio', 'Volumes Traded', 'Inventory Turnover',
       'Covid Impact (Beta)', 'Tracking Error', 'Dollar Exchange Rate',
       'Put-Call Ratio', 'P/B Ratio']]
y = df['Stock Price']
```

**Preprcessing data**
using labelencoder to convert lables(words) to numbers
applying standard scalar to standardize

```
encoder = LabelEncoder()
sv = StandardScaler()
x['Industry'] = encoder.fit_transform(x['Industry'].values)
x['Index'] = encoder.fit_transform(x['Index'].values)

x = sv.fit_transform(x)
```

Splitting the data into x and y

splitting x and y using train test split in ratio 70-30

```
x_train , x_test , y_train ,  y_test = train_test_split(x,y,test_size = 0.3)
```

**Linear Regression model**

```
model = LinearRegression()
model.fit(x_train,y_train)
pred  = model.predict(x_test)
print('Model Score', model.score(x_test,y_test))
print('Mean absolute error' ,mean_absolute_error(pred,y_test))
```

**using Artificial Neural Network**

```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense , Activation , Dropout ,Input
from tensorflow.keras.optimizers import SGD
```

```
model  = tf.keras.models.Sequential()
model.add(Input(shape =(13,)))
model.add(Dense(128,activation="relu",kernel_regularizer='L1'))
model.add(Dropout(0.2))
model.add(Dense(256,activation="relu",kernel_regularizer='L1'))
model.add(Dropout(0.2))
model.add(Dense(128,activation="relu",kernel_regularizer='L1'))
# model.add(Dropout(0.2))
model.add(Dense(1,activation="linear"))
opt = SGD(lr=0.01, momentum=0.9)
model.compile(optimizer = "Adam",
              loss = "mean_absolute_error")
```
