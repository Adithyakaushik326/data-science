```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],bins=10,kde = False)
```

```python
sns.jointplot(x='total_bill',y= 'tip',data =tips,kind = 'kde')
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.distplot(tips['tip'])
```

```python
sns.jointplot(x='tip',y='total_bill' ,data =tips,kind ='hex')
```

```python
sns.pairplot(tips)
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.barplot(x='sex',y='total_bill',data =tips)
```

```python
sns.countplot(x='sex',data = tips)
```

```python
sns.boxplot(x='day',y='total_bill',data =tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter =True,hue='sex',split=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='strip')
```
